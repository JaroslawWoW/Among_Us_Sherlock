```python
for x in range(len(all_files)):
    # Wczytaj obraz
    img = cv2.imread("D:\\Project Sherlock\\Grafiki\\PC\\{}".format(all_files[x]))
    

    # Koordynaty (przyk≈Çad)
    x1, y1 = 1800, 540
    x2, y2 = 2170, 1030

    # Wycinamy fragment
    cropped = img[y1:y2, x1:x2]

    # Szukamy czarnej obw√≥dki ‚Äî tworzymy maskƒô "ciemnych" pikseli
    # Zak≈Çadamy, ≈ºe obw√≥dka ma warto≈õci RGB bliskie 0
    lower_black = np.array([0, 0, 0])
    upper_black = np.array([50, 50, 50])
    mask_black = cv2.inRange(cropped, lower_black, upper_black)

    # Znalezienie najwiƒôkszego konturu (ca≈Ça postaƒá z obw√≥dkƒÖ)
    contours, _ = cv2.findContours(mask_black, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)

        # Tworzymy pustƒÖ maskƒô
        mask = np.zeros(cropped.shape[:2], dtype=np.uint8)

        # Wype≈Çniamy kontur (wnƒôtrze postaci + obw√≥dka) na bia≈Ço
        cv2.drawContours(mask, [largest_contour], -1, 255, thickness=cv2.FILLED)

        # Konwersja maski do kana≈Çu alfa
        alpha = mask

        # Rozdziel kana≈Çy
        b, g, r = cv2.split(cropped)

        # ≈ÅƒÖczenie w obraz z alfa
        rgba = cv2.merge([b, g, r, alpha])

        # Zapisz
        cv2.imwrite("D:\\Project Sherlock\\Grafiki\\PC\\{}_masked.png".format(names[x]), rgba)
        print("Zapisano jako {}_masked.png".format(names[x]))
    else:
        print("Nie znaleziono konturu obw√≥dki.")
```

```python
def overlay_image(background, overlay, x, y, scale_factor):
    """Osadza overlay na background w pozycji (x, y) z obs≈ÇugƒÖ alfa."""
    
    
    
    new_w = int(overlay.shape[1] * scale_factor)
    new_h = int(overlay.shape[0] * scale_factor)

    # U≈ºyj cv2.resize
    overlay = cv2.resize(overlay, (new_w, new_h), interpolation=cv2.INTER_LINEAR)


    bh, bw = background.shape[:2]
    oh, ow = overlay.shape[:2]

    if x + ow > bw or y + oh > bh:
        print("Ostrze≈ºenie: overlay wykracza poza obraz t≈Ça!")
        print(x+ow, y+oh)
        return background

    # Wyodrƒôbnij kana≈Çy
    b, g, r, a = cv2.split(overlay)
    mask = a / 255.0

    for c in range(3):  # RGB
        background[y:y+oh, x:x+ow, c] = (1.0 - mask) * background[y:y+oh, x:x+ow, c] + mask * overlay[:, :, c]

    # Nadpisujemy kana≈Ç alfa t≈Ça, je≈õli potrzebne
    if background.shape[2] == 4:
        background[y:y+oh, x:x+ow, 3] = np.maximum(background[y:y+oh, x:x+ow, 3], a)

    return background
```

```python
import random

def random_coord_in_area_with_size(area, char_width, char_height):
    """
    Losuje wsp√≥≈Çrzƒôdne w area tak, ≈ºeby ca≈Çy gracz siƒô zmie≈õci≈Ç.
    """
    width = area["x_max"] - area["x_min"]
    height = area["y_max"] - area["y_min"]

    if width < char_width or height < char_height:
        # Nie da siƒô zmie≈õciƒá ‚Äî zwr√≥ƒá None
        return None

    x = random.randint(area["x_min"], area["x_max"] - char_width)
    y = random.randint(area["y_min"], area["y_max"] - char_height)
    return (x, y)
```

```python

def spawn_players_on_map(
    map_path,
    char_path,
    areas_path,
    output_image_path,
    output_coords_path,
    num_players=1,
    char_width=51,
    char_height=690
):
    """
    Wstawia graczy na mapƒô w losowych area.
    Zapisuje nowy obraz oraz plik JSON z koordynatami.
    """

    # Wczytaj mapƒô
    map_img = cv2.imread(map_path)
    result_img = map_img.copy()

    # Wczytaj postaƒá
    char_img = cv2.imread(char_path, cv2.IMREAD_UNCHANGED)
    if char_img is None:
        raise FileNotFoundError(f"‚ùå Nie mo≈ºna wczytaƒá postaci: {char_path}")

    # Zmie≈Ñ rozmiar postaci
    char_img_resized = cv2.resize(char_img, (char_width, char_height), interpolation=cv2.INTER_AREA)

    # Wczytaj areas
    with open(areas_path, "r") as f:
        spawn_areas = json.load(f)

    coords_list = []

    for _ in range(num_players):
        # Losuj area (proporcjonalnie do powierzchni)
        area = choose_area_weighted_with_size(spawn_areas, char_width, char_height)

        if not area:
            print("‚ö†Ô∏è Nie znaleziono area. Przerywam dodawanie gracza.")
            break

        coord = random_coord_in_area_with_size(area, char_width, char_height)

        if not coord:
            print("‚ö†Ô∏è Nie uda≈Ço siƒô znale≈∫ƒá miejsca w area. Pomijam.")
            continue

        x, y = coord
        coords_list.append({"x": x, "y": y})

        # Wklej gracza na mapƒô
        if char_img_resized.shape[2] == 4:
            # Kana≈Ç alfa
            alpha_s = char_img_resized[:, :, 3] / 255.0
            alpha_l = 1.0 - alpha_s

            for c in range(3):
                result_img[y:y+char_height, x:x+char_width, c] = (
                    alpha_s * char_img_resized[:, :, c] +
                    alpha_l * result_img[y:y+char_height, x:x+char_width, c]
                )
        else:
            result_img[y:y+char_height, x:x+char_width] = char_img_resized

    # Zapisz wynikowƒÖ mapƒô
    cv2.imwrite(output_image_path, result_img)
    print(f"‚úÖ Zapisano mapƒô z graczami: {output_image_path}")

    # Zapisz wsp√≥≈Çrzƒôdne
    with open(output_coords_path, "w") as f:
        json.dump(coords_list, f, indent=4)
    print(f"‚úÖ Zapisano koordynaty do: {output_coords_path}")
```

```python
def choose_area_weighted_with_size(areas, char_width, char_height):
    filtered_areas = [
        a for a in areas
        if (a["x_max"] - a["x_min"] >= char_width) and (a["y_max"] - a["y_min"] >= char_height)
    ]

    if not filtered_areas:
        print("‚ö†Ô∏è Brak area wystarczajƒÖco du≈ºych dla tej postaci.")
        return None

    areas_sizes = [
        (a["x_max"] - a["x_min"]) * (a["y_max"] - a["y_min"])
        for a in filtered_areas
    ]
    total_size = sum(areas_sizes)
    probabilities = [size / total_size for size in areas_sizes]

    area = random.choices(filtered_areas, weights=probabilities, k=1)[0]
    return area


def random_coord_in_area_with_size(area, char_width, char_height):
    width = area["x_max"] - area["x_min"]
    height = area["y_max"] - area["y_min"]

    if width < char_width or height < char_height:
        return None

    x = random.randint(area["x_min"], area["x_max"] - char_width)
    y = random.randint(area["y_min"], area["y_max"] - char_height)
    return (x, y)
```

```python
def are_close_or_overlapping(a, b, threshold=20):
    """
    Sprawdza czy dwa prostokƒÖty (areny) sƒÖ blisko siebie lub zachodzƒÖ na siebie.
    threshold - maksymalna odleg≈Ço≈õƒá, przy kt√≥rej ≈ÇƒÖczymy
    """
    # Sprawdzenie odleg≈Ço≈õci w osi X
    horizontal_close = not (a['x_max'] + threshold < b['x_min'] or b['x_max'] + threshold < a['x_min'])
    # Sprawdzenie odleg≈Ço≈õci w osi Y
    vertical_close = not (a['y_max'] + threshold < b['y_min'] or b['y_max'] + threshold < a['y_min'])

    # Je≈õli sƒÖ blisko w obu osiach - mo≈ºna ≈ÇƒÖczyƒá
    return horizontal_close and vertical_close

def merge_two_areas(a, b):
    """
    ≈ÅƒÖczy dwie areny w jednƒÖ, biorƒÖc min/max z obu
    """
    return {
        'x_min': min(a['x_min'], b['x_min']),
        'x_max': max(a['x_max'], b['x_max']),
        'y_min': min(a['y_min'], b['y_min']),
        'y_max': max(a['y_max'], b['y_max']),
    }

def merge_close_areas(areas, threshold=20):
    """
    Scala listƒô aren, ≈ÇƒÖczƒÖc te, kt√≥re sƒÖ blisko siebie.
    """
    merged = True
    while merged:
        merged = False
        new_areas = []
        skip_indices = set()

        for i in range(len(areas)):
            if i in skip_indices:
                continue
            a = areas[i]
            merged_with_any = False
            for j in range(i + 1, len(areas)):
                if j in skip_indices:
                    continue
                b = areas[j]
                if are_close_or_overlapping(a, b, threshold):
                    # scal a i b
                    new_area = merge_two_areas(a, b)
                    new_areas.append(new_area)
                    skip_indices.update([i, j])
                    merged = True
                    merged_with_any = True
                    break
            if not merged_with_any and i not in skip_indices:
                new_areas.append(a)

        areas = new_areas
    return areas

# Przyk≈Çad u≈ºycia:
import json

with open("D:\\Project Sherlock\\spawn_areas.json", "r") as f:
    spawn_areas = json.load(f)

merged_areas = merge_close_areas(spawn_areas, threshold=20)

print(f"Liczba stref przed: {len(spawn_areas)}")
print(f"Liczba stref po scaleniu: {len(merged_areas)}")

# Mo≈ºesz teraz zapisaƒá wynik do pliku:
with open("D:\\Project Sherlock\\spawn_areas_merged.json", "w") as f:
    json.dump(merged_areas, f, indent=2)
```
